```python
import pandas as pd
```

```python
# Construye la ruta relativa al archivo 
a = "../primer_proyecto_datos/milan_airbnb.csv"
b = "../primer_proyecto_datos/london_airbnb.csv"
c = "../primer_proyecto_datos/madrid_airbnb.csv"
d = "../primer_proyecto_datos/NY_airbnb.csv"
e = "../primer_proyecto_datos/sydney_airbnb.csv"
f = "../primer_proyecto_datos/tokyo_airbnb.csv"
```

```python
# Rutas relativas
rutas = {
    'Milán': a,
    'Londres': b,
    'Madrid': c,
    'Nueva York': d,
    'Sídney': e,
    'Tokio': f
}

# Cargar los datasets, agregar la columna 'ciudad' y unirlos
dataframes = []

for ciudad, ruta in rutas.items():
    df = pd.read_csv(ruta)
    df['ciudad'] = ciudad
    dataframes.append(df)

combined_df = pd.concat(dataframes, ignore_index=True)
```

```python
# Código para guardar el DataFrame transformado en un archivo CSV usando una ruta relativa
combined_df.to_csv("../primer_proyecto_datos/combined_airbnb.csv", index=False)
```
