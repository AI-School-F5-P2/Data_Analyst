## Detallamos las variables dadas 

- **id**: Número identificador del cliente

- **host_id**: Nombre de inquilino
- **neighbourhood_group**: Grupo del barrio al que pertenece
- **neighbourhood**: Vecindario
- **latitude**: Latitud
- **longitude**: Longitud
- **room_type**: Tipo de habitación (Toda la casa/apto - Habitación privada)
- **price**: Precio
- **minimum_nights**: Cantidad de noches minimas arrendadas
- **number_of_reviews**: Número de reseñas hechas
- **last_review**: Última revisión
- **reviews_per_month**: Cantidad de reseñas hechas por mes
- **calculated_host_listings_count**: Cantidad de inquilinos
- **availability_365**: Disponibilidad al año.

```python
# Importa Pandas
import pandas as pd

# Define tu función summary (asegúrate de que la función esté definida antes de llamarla)

def summary(df):
    total_values = df.shape[0]
    missing_data = df.isnull().sum()
    missing_percentage = (missing_data / total_values) * 100
    print(f"Dataset has {df.shape[1]} features and {df.shape[0]} rows.")
    summary = pd.DataFrame(index=df.columns)
    summary["Unique"] = df.nunique().values
    summary["Missing"] = df.isnull().sum().values
    summary['Missing %'] = ((missing_data / total_values) * 100).round(2)
    summary["Duplicated"] = df.duplicated().sum()
    summary["Types"] = df.dtypes
    return summary

# Llama a la función summary y pasa el DataFrame df como argumento
resumen_datos = summary(df)

# Imprime el resumen de datos
print(resumen_datos)
```
