## Detallamos las variables dadas 

- **id**: Número identificador del cliente
- **name**: Descripción del Airbnb, puede que contengan ciertas características
- **host_id**: Nombre de inquilino
- **neighbourhood_group**: Grupo del barrio al que pertenece
- **neighbourhood**: Vecindario
- **latitude**: Latitud
- **longitude**: Longitud
- **room_type**: Tipo de habitación (Toda la casa/apto - Habitación privada)
- **price**: Precio
- **minimum_nights**: Cantidad de noches minimas arrendadas
- **number_of_reviews**: Número de reseñas hechas
- **last_review**: Última revisión
- **reviews_per_month**: Cantidad de reseñas hechas por mes
- **calculated_host_listings_count**: Cantidad de inquilinos
- **availability_365**: Disponibilidad al año.
